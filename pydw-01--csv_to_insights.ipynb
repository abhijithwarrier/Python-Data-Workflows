**PYTHON SCRIPT TO _LOAD, CLEAN, AND SUMMARISE CSV DATA USING PANDAS._ 📊🐍**

This script demonstrates a simple data workflow using pandas. It loads a CSV file, cleans basic issues like duplicates and missing values, and generates quick insights from the dataset.

---

### **📌 Step 1 — Load CSV**

Load the dataset into a pandas DataFrame.

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-01-sample_datasets.csv")
print("✅ Data Loaded")
print(df.head())
```

---

### **📌 Step 2 — Inspect Dataset**

Understand structure, columns, and data types.

```python
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
print("Data Types:\n", df.dtypes)
```

---

### **📌 Step 3 — Clean Data**

Standardise columns, remove duplicates, handle missing values.

```python
# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# Remove duplicates
df = df.drop_duplicates()

# Handle missing values
for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("Unknown")

print("✅ Data Cleaned")
```

---

### **📌 Step 4 — Summary Statistics**

Quick overview of dataset.

```python
print(df.describe(include="all"))
```

---

### **📌 Step 5 — Generate Insights**

Group data for meaningful insights.

```python
if "category" in df.columns and "sales" in df.columns:
    insights = df.groupby("category")["sales"].sum().sort_values(ascending=False)
    print("📊 Sales by Category:\n", insights)
```

---

### **📌 Key Takeaways**

- Always inspect your dataset before starting any analysis
- Cleaning (duplicates, missing values, formatting) is essential for reliable results
- Simple summaries help you quickly understand the data
- Grouping data reveals meaningful patterns and insights

From raw CSV to insights — this is how every data workflow begins.
