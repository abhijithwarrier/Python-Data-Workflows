**PYTHON SCRIPT TO _BUILD A DATA CLEANING PIPELINE (MISSING VALUES, DUPLICATES, TYPES)_ 🧹📊**

This script demonstrates a structured data cleaning pipeline.

It handles missing values, removes duplicates, and standardises data types for reliable analysis.

---

### **📌 Step 1 — Load Dataset**

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-02--sample_datasets.csv")
print("✅ Data Loaded")
```

---

### **📌 Step 2 — Inspect Missing Values**

```python
print("📌 Missing Values:")
print(df.isnull().sum())
```

---

### **📌 Step 3 — Handle Missing Values**

```python
for col in df.columns:
    if df[col].dtype in ["int64", "float64"]:
        df[col] = df[col].fillna(df[col].median())
    else:
        df[col] = df[col].fillna("Unknown")

print("✅ Missing values handled")
```

---

### **📌 Step 4 — Remove Duplicates**

```python
before = len(df)
df = df.drop_duplicates()
after = len(df)

print(f"🧹 Removed {before - after} duplicate rows")
```

---

### **📌 Step 5 — Fix Data Types**

```python
# Convert date column
if "order_date" in df.columns:
    df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

# Ensure numeric columns
for col in ["sales", "quantity"]:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

print("✅ Data types standardized")
```

---

### **📌 Step 6 — Verify Clean Data**

```python
print("📊 Final Data Info:")
print(df.info())
```

---

### **📌 Key Takeaways**

- Missing values should be handled based on data type
- Removing duplicates prevents skewed analysis
- Correct data types are critical for processing and aggregation
- A structured pipeline makes data cleaning repeatable
