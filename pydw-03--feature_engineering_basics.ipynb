**PYTHON SCRIPT TO _CREATE NEW FEATURES AND TRANSFORM DATA USING PANDAS_ 🧩📊**

This script demonstrates basic feature engineering using pandas. It creates new columns, transforms existing data, and prepares datasets for better analysis.

---

### **📌 Step 1 — Load Dataset**

Load the CSV dataset using pandas.

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-03--sample_datasets.csv")

print("✅ Data Loaded")
print(df.head())
```

---

### **📌 Step 2 — Fix Basic Data Types**

Convert date and numeric columns before creating features.

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")

print("✅ Data types fixed")
```

---

### **📌 Step 3 — Create Revenue Per Item**

Create a new feature from existing numeric columns.

```python
df["revenue_per_item"] = df["sales"] / df["quantity"]

print(df[["sales", "quantity", "revenue_per_item"]].head())
```

----

### **📌 Step 4 — Extract Date Features**

Create useful columns from the order date.

```python
df["order_year"] = df["order_date"].dt.year
df["order_month"] = df["order_date"].dt.month
df["order_day"] = df["order_date"].dt.day
df["order_weekday"] = df["order_date"].dt.day_name()

print(df[["order_date", "order_year", "order_month", "order_weekday"]].head())
```

---

### **📌 Step 5 — Create Sales Category**

Convert raw sales values into simple business labels.

```python
def sales_category(sales):
    if sales >= 2000:
        return "High"
    elif sales >= 1000:
        return "Medium"
    else:
        return "Low"


df["sales_category"] = df["sales"].apply(sales_category)

print(df[["sales", "sales_category"]].head())
```

---

### **📌 Step 6 — Encode Category Column**

Convert text categories into numeric codes.

```python
df["category_code"] = df["category"].astype("category").cat.codes

print(df[["category", "category_code"]].head())
```

---

### **📌 Step 7 — Save Transformed Dataset**

Save the new feature-engineered dataset.

```python
df.to_csv("feature_engineered_data.csv", index=False)

print("💾 Feature engineered dataset saved")
```

---

### **📌 Key Takeaways**

- Feature engineering helps create more useful information from raw data
- Date columns can be transformed into year, month, day, and weekday features
- Numeric columns can be combined to create business-friendly metrics
- Text categories can be converted into numeric values for analysis or ML

---

### **🚀 Final Note**

Feature engineering is where raw data starts becoming more meaningful for analysis, dashboards, and machine learning.
