**PYTHON SCRIPT TO _BUILD A SIMPLE DATA DASHBOARD USING MATPLOTLIB & SEABORN_ 📊📈**

This script demonstrates how to visualise cleaned data. It creates simple charts to understand trends, categories, and distributions.

---

### **📌 Step 1 — Load Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("datasets/pydw-04--sample_datasets.csv")

print("✅ Data Loaded")
```

---

### **📌 Step 2 — Basic Data Preparation**

Ensure correct data types for visualisation.

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["sales"] = pd.to_numeric(df["sales"], errors="coerce")

df = df.dropna(subset=["order_date", "sales"])
```

---

### **📌 Step 3 — Sales by Category (Bar Chart)**

```python
plt.figure(figsize=(8, 5))

category_sales = df.groupby("category")["sales"].sum().sort_values()

category_sales.plot(kind="bar")

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.tight_layout()
plt.show()
```

---

### **📌 Step 4 — Sales Trend Over Time (Line Chart)**

```python
plt.figure(figsize=(8, 5))

daily_sales = df.groupby("order_date")["sales"].sum()

daily_sales.plot()

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Sales")

plt.tight_layout()
plt.show()
```

---

### **📌 Step 5 — Sales Distribution (Histogram)**

```python
plt.figure(figsize=(8, 5))

sns.histplot(df["sales"], bins=10, kde=True)

plt.title("Sales Distribution")

plt.tight_layout()
plt.show()
```

---

### **📌 Step 6 — Sales by Region (Seaborn Barplot)**

```python
plt.figure(figsize=(8, 5))

sns.barplot(data=df, x="region", y="sales", estimator=sum)

plt.title("Sales by Region")

plt.tight_layout()
plt.show()
```

---

### **📌 Key Takeaways**

- Visualisations help quickly understand patterns in data
- Bar charts are useful for comparing categories
- Line charts help identify trends over time
- Histograms show data distribution
- Simple dashboards can be built with just pandas + matplotlib + seaborn
