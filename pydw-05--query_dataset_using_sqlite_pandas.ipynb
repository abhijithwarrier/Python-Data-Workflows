**PYTHON SCRIPT TO _QUERY DATA USING SQLITE & PANDAS_ 🗄️📊**

This script demonstrates how to combine SQL and Python for data analysis. It loads CSV data into SQLite and runs SQL queries using pandas.

---

### **📌 Step 1 — Import Libraries**

Import pandas and sqlite3.

```python
import pandas as pd
import sqlite3
```

---

### **📌 Step 2 — Load Dataset**

Load the CSV dataset into pandas.

```python
df = pd.read_csv("datasets/pydw-05--sample_datasets.csv")

print("✅ Dataset Loaded")
print(df.head())
```

---

### **📌 Step 3 — Create SQLite Database**

Create a local SQLite database connection.

```python
conn = sqlite3.connect("sales_data.db")

print("✅ SQLite database connected")
```

---

### **📌 Step 4 — Store Data in SQLite Table**

Save the DataFrame into a SQL table.

```python
df.to_sql("sales", conn, if_exists="replace", index=False)

print("✅ Dataset stored in SQLite table")
```

---

### **📌 Step 5 — Query Total Sales by Category**

Run SQL queries using pandas.

```python
query = """
SELECT
    category,
    SUM(sales) AS total_sales
FROM sales
GROUP BY category
ORDER BY total_sales DESC
"""

result = pd.read_sql(query, conn)

print(result)
```

---

### **📌 Step 6 — Query Regional Sales**

Analyze sales by region.

```python
query = """
SELECT
    region,
    AVG(sales) AS avg_sales
FROM sales
GROUP BY region
ORDER BY avg_sales DESC
"""

region_result = pd.read_sql(query, conn)

print(region_result)
```

---

### **📌 Step 7 — Query Top Orders**

Fetch highest sales orders.

```python
query = """
SELECT
    order_id,
    category,
    region,
    sales
FROM sales
ORDER BY sales DESC
LIMIT 5
"""

top_orders = pd.read_sql(query, conn)

print(top_orders)
```

---

### **📌 Step 8 — Close Database Connection**

```python
conn.close()

print("🔒 Database connection closed")
```

---

### **📌 Key Takeaways**

- SQLite allows lightweight local database analysis
- pandas can directly execute SQL queries
- SQL is powerful for aggregation and filtering
- Combining SQL + Python creates flexible data workflows

---

### **🚀 Final Note**

SQL and Python together make data analysis faster, cleaner, and more scalable.
