**PYTHON SCRIPT TO _FETCH DATA FROM AN API AND ANALYZE IT USING PANDAS_. 🌐📊**

This script demonstrates how to fetch data from a public API. It loads API responses into pandas and performs basic analysis to extract insights.

---

### **📌 Step 1 — Import Libraries**

Import the required libraries.

```python
import requests
import pandas as pd
```

---

### **📌 Step 2 — Fetch Data from API**

Retrieve data from a public API endpoint.

```python
url = "https://jsonplaceholder.typicode.com/posts"

response = requests.get(url)
data = response.json()

print(f"✅ Retrieved {len(data)} records")
```

---

### **📌 Step 3 — Convert API Response to DataFrame**

Load the JSON response into pandas.

```python
df = pd.DataFrame(data)

print(df.head())
```

---

### **📌 Step 4 — Inspect Dataset**

Explore the structure of the API data.

```python
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nData Types:\n", df.dtypes)
```

---

### **📌 Step 5 — Analyze Posts Per User**

Group data by user.

```python
posts_per_user = (
    df.groupby("userId")
      .size()
      .reset_index(name="total_posts")
)

print(posts_per_user)
```

---

### **📌 Step 6 — Find Longest Post Titles**

Create a new feature for title length.

```python
df["title_length"] = df["title"].str.len()

top_titles = (
    df.sort_values("title_length", ascending=False)
      [["id", "title", "title_length"]]
      .head(5)
)

print(top_titles)
```

---

### **📌 Step 7 — Export Results**

Save processed data for later use.

```python
df.to_csv("datasets/output/pydw-06--sample_output_datasets.csv", index=False)

print("💾 Data exported successfully")
```

---

### **📌 Key Takeaways**

- APIs are a common source of real-world data
- JSON responses can be converted directly into DataFrames
- Grouping and transformations help extract insights
- API data can be integrated into larger workflows
