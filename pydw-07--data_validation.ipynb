**PYTHON SCRIPT TO VALIDATE DATA QUALITY USING CUSTOM RULES ✅📊**

This script demonstrates how to validate datasets before analysis. It checks for missing values, invalid records, duplicate entries, and rule violations to ensure data quality.

---

### **📌 Step 1 — Load Dataset**

Load the CSV dataset using pandas.

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-06--sample_datasets.csv")

print("✅ Dataset Loaded")
print(df.head())
```

---

### **📌 Step 2 — Check Missing Values**

Identify columns containing null values.

```python
missing_values = df.isnull().sum()

print("📌 Missing Values:")
print(missing_values)
```

---

### **📌 Step 3 — Check Duplicate Records**

Detect duplicate rows.

```python
duplicates = df.duplicated().sum()

print(f"📌 Duplicate Records: {duplicates}")
```

---

### **📌 Step 4 — Validate Sales Values**

Sales should never be negative.

```python
invalid_sales = df[df["sales"] < 0]

print("📌 Invalid Sales Records:")
print(invalid_sales)
```

---

### **📌 Step 5 — Validate Quantity Values**

Quantity should be greater than zero.

```python
invalid_quantity = df[df["quantity"] <= 0]

print("📌 Invalid Quantity Records:")
print(invalid_quantity)
```

---

### **📌 Step 6 — Validate Category Values**

Ensure category belongs to approved values.

```python
valid_categories = [
    "Electronics",
    "Furniture",
    "Clothing"
]

invalid_categories = df[
    ~df["category"].isin(valid_categories)
]

print("📌 Invalid Categories:")
print(invalid_categories)
```

---

### **📌 Step 7 — Generate Validation Summary**

Create a quick validation report.

```python
validation_report = {
    "missing_values": int(df.isnull().sum().sum()),
    "duplicates": int(df.duplicated().sum()),
    "invalid_sales": len(invalid_sales),
    "invalid_quantity": len(invalid_quantity),
    "invalid_categories": len(invalid_categories)
}

print(validation_report)
```

---

### **📌 Key Takeaways**

- Validation helps identify bad data before analysis
- Business rules improve dataset reliability
- Invalid records should be flagged early in the workflow
- Data quality checks prevent inaccurate insights

---

### **🚀 Final Note**

Good analysis starts with trusted data, and trusted data starts with validation.
