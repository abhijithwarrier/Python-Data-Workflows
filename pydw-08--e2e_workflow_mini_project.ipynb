**PYTHON SCRIPT FOR _AN END-TO-END DATA WORKFLOW USING PANDAS_ 📊🚀**

This mini project demonstrates a complete data workflow. It loads data, cleans it, validates records, creates features, generates insights, and exports a final report.

---

### **📌 Step 1 — Load Dataset**

Load the sales dataset into pandas.

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-08--sample_datasets.csv")

print("✅ Dataset Loaded")
print(df.head())
```

---

### **📌 Step 2 — Clean the Data**

Handle missing values and remove duplicates.

```python
df = df.drop_duplicates()

df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")

df["sales"] = df["sales"].fillna(df["sales"].median())
df["quantity"] = df["quantity"].fillna(df["quantity"].median())

print("✅ Data Cleaned")
```

---

### **📌 Step 3 — Validate Records**

Identify invalid sales values.

```python
invalid_sales = df[df["sales"] < 0]

print("Invalid Records Found:")
print(invalid_sales)
```

---

### **📌 Step 4 — Create Features**

Generate new business metrics.

```python
df["revenue_per_item"] = df["sales"] / df["quantity"]

df["sales_category"] = df["sales"].apply(
    lambda x: "High" if x >= 2000 else
              "Medium" if x >= 1000 else
              "Low"
)

print("✅ Features Created")
```

---

### **📌 Step 5 — Generate Insights**

Analyze category performance.

```python
category_summary = (
    df.groupby("category")
      .agg(
          total_sales=("sales", "sum"),
          avg_sales=("sales", "mean"),
          total_orders=("order_id", "count")
      )
      .reset_index()
)

print(category_summary)
```

---

### **📌 Step 6 — Find Top Performing Region**

Identify the highest-selling region.

```python
region_summary = (
    df.groupby("region")["sales"]
      .sum()
      .sort_values(ascending=False)
)

print(region_summary)
```

---

### **📌 Step 7 — Export Reports**

Save processed data and summaries.

```python
df.to_csv("datasets/output/processed_sales_data.csv", index=False)

category_summary.to_csv(
    "category_summary.csv",
    index=False
)

print("💾 Reports Exported")
```

---

### **📌 Key Takeaways**

- A complete workflow involves multiple stages, not just analysis
- Cleaning and validation improve data reliability
- Feature engineering creates additional business value
- Insights become more meaningful when built on quality data
