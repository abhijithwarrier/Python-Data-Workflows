**PYTHON SCRIPT TO _BUILD A SIMPLE ETL (EXTRACT → TRANSFORM → LOAD) PIPELINE_ 🔄📊**

This script demonstrates a basic ETL workflow using Python and pandas. It extracts data from a CSV file, transforms it through cleaning and feature engineering, and loads the processed data into a new CSV.

---

### **📌 Step 1 — Extract Data**

Read the source dataset.

```python
import pandas as pd

df = pd.read_csv("datasets/pydw-09--sample_datasets.csv")

print("Data Extracted")
print(df.head())
```

---

### **📌 Step 2 — Inspect Dataset**

Understand the incoming data.

```python
print(df.shape)
print(df.dtypes)
print(df.isnull().sum())
```

---

### **📌 Step 3 — Transform Data**

Clean and standardize the dataset.

```python
df = df.drop_duplicates()

df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")

df["sales"] = df["sales"].fillna(df["sales"].median())
df["quantity"] = df["quantity"].fillna(df["quantity"].median())
```

---

### **📌 Step 4 — Create New Features**

Generate business-friendly columns.

```python
df["revenue_per_item"] = df["sales"] / df["quantity"]

df["sales_category"] = df["sales"].apply(
    lambda x: "High"
    if x >= 2000
    else "Medium"
    if x >= 1000
    else "Low"
)
```

---

### **📌 Step 5 — Generate Summary**

Summarize transformed data.

```python
summary = (
    df.groupby("category")
      .agg(
          total_sales=("sales", "sum"),
          average_sales=("sales", "mean"),
          total_orders=("order_id", "count")
      )
      .reset_index()
)

print(summary)
```

---

### **📌 Step 6 — Load Processed Data**

Export the transformed data and summary.

```python
df.to_csv("datasets/output/pydw-09--processed_sales_data.csv", index=False)

summary.to_csv("datasets/output/pydw-09--sales_summary.csv", index=False)

print("ETL Pipeline Completed")
```

---

### **📌 Key Takeaways**

- ETL stands for Extract, Transform, and Load
- Data is cleaned before analysis
- Transformations add value to raw datasets
- Processed datasets are ready for reporting and analytics
